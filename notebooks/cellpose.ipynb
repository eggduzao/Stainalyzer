## Jupyter Notebook: Cellpose Testing

This notebook will explore Cellpose.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from cellpose import models
```

```python
# ðŸ“‚ Load your image
prefix = "/Users/egg/Projects/Stainalyzer/data/"
image_path = os.path.join(prefix, "interim/images_to_enhance/cell_00058.png")
img = imread(image_path)
```

```python
# Check image shape
print(f"Image shape: {img.shape}, dtype: {img.dtype}")
```

```python
# Choose a Cellpose pre-trained model
model = models.Cellpose(model_type="cyto")  # or "nuclei" for nuclear segmentation
```

```python
# Run segmentation
masks, flows, styles, diams = model.eval(img, diameter=None, channels=[0, 0])
```

```python
# Display results
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

# Segmentation mask
plt.subplot(1, 2, 2)
plt.imshow(masks, cmap="jet")
plt.title("Cellpose Segmentation")
plt.axis("off")

plt.show()
```
